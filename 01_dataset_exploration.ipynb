```python
 import os
import scipy.io as sio
import matplotlib.pyplot as plt

base_path = r"C:\Users\sreej\Downloads\ShanghaiTech\part_A\train_data"
img_path = os.path.join(base_path, "images")
gt_path = os.path.join(base_path, "ground-truth")  # <-- correct folder name

# List sample files
print("Sample images:", os.listdir(img_path)[:5])
print("Sample ground truth files:", os.listdir(gt_path)[:5])

# Load the first image's ground truth
images = os.listdir(img_path)
images.sort()  # Ensure they match the GT files in order
mat_file = os.path.join(gt_path, "GT_" + images[0].split('.')[0] + ".mat")

gt = sio.loadmat(mat_file)
density_map = gt["image_info"][0][0][0][0][0]

plt.imshow(density_map, cmap="jet")
plt.show()
```
