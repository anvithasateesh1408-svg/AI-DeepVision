```python
import os
from pathlib import Path
import cv2
import matplotlib.pyplot as plt
import numpy as np

DATA_ROOT = Path(r"D:\chromedocuments\DeepVision Crowd Monitor AI\ShanghaiTech")
print("Dataset exists:", DATA_ROOT.exists())
```

```python
parts = ["part_A", "part_B"]

for p in parts:
    train_dir = DATA_ROOT / p / "train_data" / "images"
    test_dir = DATA_ROOT / p / "test_data" / "images"
    print(f"\n{p.upper()} :")
    print(" Train images:", len(list(train_dir.glob("*.jpg"))))
    print(" Test images :", len(list(test_dir.glob("*.jpg"))))
```

```python
def show_images(img_paths, n=6):
    plt.figure(figsize=(12,8))
    for i, p in enumerate(img_paths[:n]):
        img = cv2.imread(str(p))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.subplot(2,3,i+1)
        plt.imshow(img)
        plt.axis('off')
    plt.tight_layout()
    plt.show()

sample_imgs = list((DATA_ROOT / "part_A" / "train_data" / "images").glob("*.jpg"))
show_images(sample_imgs, n=6)
```

```python
img_dir = DATA_ROOT / "part_A" / "train_data" / "images"

sizes = []
corrupted = []

for imgp in img_dir.glob("*.jpg"):
    img = cv2.imread(str(imgp))
    if img is None:
        corrupted.append(imgp.name)
    else:
        sizes.append(img.shape[:2])

print("Total images:", len(sizes))
print("Corrupted:", corrupted[:10])
print("Sample sizes:", sizes[:5])
```
